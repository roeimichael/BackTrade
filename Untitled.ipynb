```python
import numpy as np
import random
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import glob
```

```python
df = pd.read_csv('./data/stocks/AAPL.csv')
print("size of data frame: ", df.shape)
cols = df.columns
for col in cols:
    if "CDL" not in col:
        print(col)
        plt.hist(df[col], bins = 100)
        plt.show()
```

```python
for col in cols:
    if "CDL" not in col and "Date" not in col and "ticker" not in col:
        print(col)
        plt.boxplot(df[col])
        plt.show()
```
